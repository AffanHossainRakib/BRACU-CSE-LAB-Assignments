#String

```python
#@title Task 1
var = input()
upper_case = 0
lower_case = 0
final_string_u = ""
final_string_l = ""

for i in var:
    if ord("A") <= ord(i) <= ord("Z"):
        upper_case +=1
        final_string_u += i
        final_string_l += chr(ord(i)+32)
    elif ord("a") <= ord(i) <= ord("z"):
        lower_case += 1
        final_string_u += chr(ord(i)-32)
        final_string_l += i

if upper_case > lower_case:
    print(final_string_u)
else:
    print(final_string_l)
```

```python
#@title Task 2
var = input()
v_words = 0
v_numbers = 0
for i in var:
    if ord("A") <= ord(i) <= ord("Z") or ord("a") <= ord(i) <= ord("z"):
        v_words += 1
    elif ord("0") <= ord(i) <= ord("9"):
        v_numbers += 1
if v_words == 0:
    print("NUMBER")
elif v_numbers == 0:
    print("WORD")
else:
    print("MIXED")
```

```python
#@title Task 3
var = input()
# f_str = ""
index_1 = 0
index_2 = 0
for i in range(len(var)):
    if ord("A") <= ord(var[i]) <= ord("Z"):
        index_1 = i
        break
for i in range(len(var)-1,-1,-1):
    if ord("A") <= ord(var[i]) <= ord("Z"):
        index_2 = i
        break
f_str = var[index_1+1:index_2:1]
if f_str == "":
    print("BLANK")
else:
    print(f_str)
```

```python
#@title Task 4
word1, word2 = input().split(",")
f_str = ""
for i in word1:
    if i in word2:
        f_str += i
for i in word2:
    if i in word1:
        f_str += i
if f_str != "":
    print(f_str)
else:
    print("Nothing in common")
```

```python
#@title Task 5
var = input()
special_c = "_$#@"
lower = False
upper = False
digit = False
special = False
f_str = ""
for i in var:
    if ord("a") <= ord(i) <= ord("z"):
        lower = True
    elif ord("A") <= ord(i) <= ord("Z"):
        upper = True
    elif ord("0") <= ord(i) <= ord("9"):
        digit = True
    elif i in special_c:
        special = True
if (lower == True) and (upper == True) and (digit == True) and (special == True):
    f_str = "OK"
else:
    if lower == False:
        f_str += "Lowercase character missing,"
    if upper == False:
        f_str += "Uppercase character missing,"
    if digit == False:
        f_str += "Digit missing,"
    if special == False:
        f_str += "Special character missing,"
if f_str == "OK":
    print(f_str)
else:
    print(f_str[0:-1:1])
```

#List

```python
#@title Task 1
myList = []
myList1 = []
while True:
    var = input()
    if var == "STOP":
        break
    else:
        myList.append(var)
        if var not in myList1:
            myList1.append(var)

for i in myList1:
    counter = 0
    for j in myList:
        if i == j:
            counter += 1
    print(f"{i} - {counter} times")
```

```python
#@title Task 2
t = int(input())
maximum = 0

for i in range(t):
    result = 0
    var = input().split()

    for j in var:
        result += int(j)

    if result >= maximum:
        maximum = result
        f_list = []
        for k in var:
            f_list.append(int(k))

print(maximum)
print(f_list)
```

```python
#@title Task 4
while True:
    user = input()
    var = user.split()
    n = len(var)
    if user == "STOP":
        break
    else:
        flag = True
        for j in range(n-1):
            if abs(int(var[j]) - int(var[j+1])) >= n:
                flag = False
                break
        if flag == True:
            print("UB Jumper")
        else:
            print("Not UB Jumper")
```

```python
#@title Task 5
n, k = input().split()
var = input().split()
eligible = []
team = []
counter = 0

for i in var:
    if int(i) < 5:
        eligible.append(int(i))
eligible.sort()

for i in range(0,int(n),3):
    temp = eligible[i:i+3:1]
    team.append(temp)

for i in team:
    if len(i) == 3:
        if i[2] <= (5-int(k)):
            counter += 1
            
print(counter)

#------------CHAT GPT-------------#
n, k = input().split()
user = input().split()

myList = []
for i in user:
    myList.append(int(i))
myList.sort()

counter = 0
for i in range(int(n)):
    if myList[i] + int(k) <= 5:
        counter += 1
    else:
        break
print(counter//3)
```

#Dictionary & Tuple

```python
#@title Task 1
given1 = {"a": 100, "b": 100, "c": 200, "d": 300}
given2 = {"a": 300, "b": 200, "d": 400, "e": 200}
newdict = {}

for i in given1:
    if i not in newdict:
        newdict[i] = given1[i]
    else:
        newdict[i] = newdict[i] + given1[i]
for i in given2:
    if i not in newdict:
        newdict[i] = given2[i]
    else:
        newdict[i] = newdict[i] + given2[i]

myList = []
for i in newdict:
    if newdict[i] not in myList:
        myList.append(newdict[i])
myList.sort()
x = tuple(myList)

print(newdict)
print(f"Values:{x}")
```

```python
#@title Task 2
mydict = {}
while True:
    var = input()
    if var == "STOP":
        break
    if var not in mydict:
        mydict[var] = 1
    else:
        mydict[var] = mydict[var] + 1
for key in mydict:
    print(f"{key} - {mydict[key]} times")
```

```python
#@title Task 3
given = {"key1" : "value1", "key2" : "value2", "key3" : "value1"}
newDict = {}
for key in given:
    if given[key] not in newDict:
        newDict[given[key]] = [key]
    else:
        newDict[given[key]].append(key)
print(newDict)
```

```python
#@title Task 4
myDict = {
    "1": (".",",","?","!",":"),
    "2": ("A","B","C"),
    "3": ("D","E","F"),
    "4": ("G","H","I"),
    "5": ("J","K","L"),
    "6": ("M","N","O"),
    "7": ("P","Q","R","S"),
    "8": ("T","U","V"),
    "9": ("W","X","Y","Z"),
    "0": (" ")
    }
sentence = input()

for character in sentence:
    for key in myDict:
        if character in myDict[key]:
            for position in range(len(myDict[key])):
                if character == myDict[key][position]:
                    print(key*(position+1),end="")
```
