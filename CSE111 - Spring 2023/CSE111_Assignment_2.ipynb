```python
#@title Task 1
def fraction(x,y):
    if y == 0 or x == 0:
        result = 0
    else:
        result = (x / y) - (x//y)
    return result
print(fraction(5,2))
print(fraction(5,0))
print(fraction(0,5))
```

```python
#@title Task 2
def BMI(height,weight):
    height = height/100
    bmi = weight/(height**2)

    if bmi < 18.5:
        condition = "Underweight"
    elif 18.5 <= bmi <= 24.9:
        condition = "Normal"
    elif 25 <= bmi <=30:
        condition = "Overweight"
    elif 30 < bmi:
        condition = "Obese"
    result = f"Score is {bmi:.1f}. You are {condition}".format(bmi)
    return result

print(BMI(175, 96))
print(BMI(152, 48))
```

```python
#@title Task 3
def khoj(x,y,z):
    result = 0
    for i in range(x,y):
        if i % z == 0:
            result += i
    return result
print(khoj(0,10,2))
print(khoj(3,16,3))
```

```python
#@title Task 4
def khuda_lagse(burger, destination="Mohakhali"):
    menu = {
        "BBQ Chicken Cheese Burger": 250,
        "Beef Burger": 170,
        "Naga Drums": 200
    }
    if destination == "Mohakhali":
        delivery = 40
    else:
        delivery = 60
    total = menu[burger] + delivery + menu[burger] * .08
    return total


print(khuda_lagse("Beef Burger", "Dhanmondi"))
print(khuda_lagse("Beef Burger"))
```

```python
#@title Task 5
def replace_domain(mail, previous = "sheba.xyz", new = "kaaj.com"):
    mail = mail.split("@")
    if mail[1] == previous:
        print(f"Unchanged: {mail[0]}@{previous}")
    else:
        print(f"Changed: {mail[0]}@{previous}")

replace_domain("alice@kaaj.com", "sheba.xyz", "kaaj.com")
replace_domain("bob@sheba.xyz", "sheba.xyz")
```

```python
#@title Task 6
def find_vowel(user):
    vowels = "AEIOUaeiou"
    wanted = []
    for c in user:
        if c in vowels:
            wanted.append(c)
    f_str = ""
    for i in wanted:
        f_str += i + ", "
    f_str = f_str[0:-2:1] + "."
    if len(wanted) == 0:
        print("No vowels in the name")
    else:
        print(f"Vowels: {f_str} Total number of vowels: {len(wanted)} ")
        
find_vowel("Steve Jobs")
find_vowel("XYZ")
```

```python
#@title Task 7
def find_palindrome(user):
    var = ""
    for i in user:
        if i != " ":
            var += i

    palindrome = False
    n = len(var)//2 - 1
    for i in range(n):
        if var[i] == var[len(var)-i-1]:
            palindrome = True
        else:
            palindrome = False
            break
    if palindrome == True:
        print("Palindrome")
    else:
        print("Not a palindrome")

find_palindrome("madam")
find_palindrome("hello")
find_palindrome("nurses run")
```

```python
#@title Task 8
def ymd (day):
    year = day//365
    month = (day - year*365) // 30
    day = (day - year*365 - month*30)
    print(f"{year} years, {month} months and {day} days")
ymd(4320)
ymd(4000)
```
