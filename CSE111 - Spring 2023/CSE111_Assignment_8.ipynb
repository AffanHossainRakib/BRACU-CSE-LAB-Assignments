```python
#@title Task 1
class RealNumber:
    
    def __init__(self, r=0):
        self.__realValue = r  
    def getRealValue(self):
        return self.__realValue
    def setRealValue(self, r):
        self.__realValue = r
    def __str__(self):
        return 'RealPart: '+str(self.getRealValue()) 
    
class ComplexNumber(RealNumber):
    def __init__(self, r=1, i=1):
        self.r = float(r)
        self.i = float(i)
        super().setRealValue(self.r)
        
    def __str__(self):
        return f"RealPart: {super().getRealValue()}\nImaginaryPart: {self.i}"

cn1 = ComplexNumber()
print(cn1)
print('---------')
cn2 = ComplexNumber(5,7)
print(cn2)
```

```python
# @title Task 2
class RealNumber:
    def __init__(self, number=0):
        self.number = number

    def __add__(self, anotherRealNumber):
        return self.number + anotherRealNumber.number

    def __sub__(self, anotherRealNumber):
        return self.number - anotherRealNumber.number

    def __str__(self):
        return str(self.number)


class ComplexNumber(RealNumber):
    def __init__(self, r, i):
        if type(r) is int:
            super().__init__(r)
        else:
            super().__init__(r.number)
        self.i = i

    def __add__(self, other):
        return ComplexNumber(super().__add__(other), self.i + other.i)
    
    def __sub__(self, other):
        return ComplexNumber(super().__sub__(other), self.i - other.i)

    def __str__(self):
        if self.i > 0:
            return f"{self.number} + {self.i}i"
        else:
            return f"{self.number} - {(-1) * self.i}i"


r1 = RealNumber(3)
r2 = RealNumber(5)
print(r1 + r2)
cn1 = ComplexNumber(2, 1)
print(cn1)
cn2 = ComplexNumber(r1, 5)
print(cn2)
cn3 = cn1 + cn2
print(cn3)
cn4 = cn1 - cn2
print(cn4)
```

```python
#@title Task 5
class Exam:
    def __init__(self, marks):
        self.marks = marks
        self.time = 60

    def examSyllabus(self):
        return "Maths , English"

    def examParts(self):
        return "Part 1 - Maths\nPart 2 - English\n"


class ScienceExam(Exam):
    def __init__(self, marks, time, *subs):
        super().__init__(marks)
        self.time = time
        self.subs = subs

    def examSyllabus(self):
        res = ""
        for i in range(len(self.subs)):
            if i == len(self.subs) - 1:
                res += self.subs[i]
            else:
                res += self.subs[i] + " , "
        return f"{super().examSyllabus()} , {res}"

    def examParts(self):
        res = ""
        for i in range(len(self.subs)):
            if i == 0:
                res += f"Part {i + 3} - {self.subs[i]}"
            else:
                res += f"\nPart {i + 3} - {self.subs[i]}"
        return f"{super().examParts()}{res}"

    def __str__(self):
        return f"Marks: {self.marks} Time: {self.time} minutes Number of Parts: {len(self.subs) + 2}"


engineering = ScienceExam(100, 90, "Physics", "HigherMaths")
print(engineering)
print('----------------------------------')
print(engineering.examSyllabus())
print(engineering.examParts())
print('==================================')
architecture = ScienceExam(100, 120, "Physics", "HigherMaths", "Drawing")
print(architecture)
print('----------------------------------')
print(architecture.examSyllabus())
print(architecture.examParts())
```

```python
#@title Task 6
class Shape3D:
    pi = 3.14159

    def __init__(self, name='Default', radius=0):
        self._area = 0
        self._name = name
        self._height = 'No need'
        self._radius = radius

    def calc_surface_area(self):
        return 2 * Shape3D.pi * self._radius

    def __str__(self):
        return "Radius: " + str(self._radius)


class Sphere(Shape3D):
    def __init__(self, name, r):
        super().__init__(name, r)
        print(f"Shape name: {self._name}, Area Formula: 4 * pi * r * r")

    def calc_surface_area(self):
        self._area = 4 * Shape3D.pi * self._radius ** 2

    def __str__(self):
        # return str("Area: ") + str(self.area)
        return f"Radius: {self._radius}, Height: {self._height}\nArea: {self._area}"


class Cylinder(Shape3D):
    def __init__(self, name, r, h):
        super().__init__(name, r)
        self._height = h
        print(f"Shape name: {self._name}, Area Formula: 2 * pi * r * (r + h)")

    def calc_surface_area(self):
        self._area = 2 * Shape3D.pi * self._radius * (self._radius + self._height)

    def __str__(self):
        return f"Radius: {self._radius}, Height: {self._height}\nArea: {self._area}"


sph = Sphere('Sphere', 5)
print('----------------------------------')
sph.calc_surface_area()
print(sph)
print('==================================')
cyl = Cylinder('Cylinder', 5, 10)
print('----------------------------------')
cyl.calc_surface_area()
print(cyl)
```

```python
#@title Task 7
class PokemonBasic:

  def __init__(self, name = 'Default', hp = 0, weakness = 'None', type = 'Unknown'):
    self.name = name
    self.hit_point = hp
    self.weakness = weakness
    self.type = type

  def get_type(self):
    return 'Main type: ' + self.type

  def get_move(self):
    return 'Basic move: ' + 'Quick Attack'

  def __str__(self):
    return "Name: " + self.name + ", HP: " + str(self.hit_point) + ", Weakness: " + self.weakness

class PokemonExtra(PokemonBasic):
    def __init__(self,  name = 'Default', hp = 0, weakness = 'None', type = 'Unknown', other=None, args=None):
        super().__init__(name, hp, weakness, type)
        self.secondType = other
        self.otherMoves = args
    def get_type(self):
        if self.secondType != None:
            return f"{super().get_type()}, Secondary type: {self.secondType}"
        else:
            return super().get_type()

    def get_move(self):
        if self.otherMoves != None:
            res= ""
            for i in range(len(self.otherMoves)):
                res += f"{self.otherMoves[i]}, "
            return f"{super().get_move()} \nOther move: {res[0:-2:1]}"
        else:
            return super().get_move()

print('\n------------Basic Info:--------------')
pk = PokemonBasic()
print(pk)
print(pk.get_type())
print(pk.get_move())

print('\n------------Pokemon 1 Info:-------------')
charmander = PokemonExtra('Charmander', 39, 'Water', 'Fire')
print(charmander)
print(charmander.get_type())
print(charmander.get_move())

print('\n------------Pokemon 2 Info:-------------')
charizard = PokemonExtra('Charizard', 78, 'Water', 'Fire', 'Flying', ('Fire Spin', 'Fire Blaze'))
print(charizard)
print(charizard.get_type())
print(charizard.get_move())

```

```python
#@title Task 10
class Department:
    def __init__(self, s):
        self.semester = s
        self.name = "Default"
        self.id = -1

    def student_info(self):
        print("Name:", self.name)
        print("ID:", self.id)

    def courses(self, c1, c2, c3):
        print("No courses Approved yet!")


class CSE(Department):
    def __init__(self, name, id, sem):
        super().__init__(sem)
        self.name = name
        self.id = id

    def courses(self, *args):
        print(f"Courses Approved to this CSE student in {self.semester} semester :")
        for i in args:
            print(i)


class EEE(Department):
    def __init__(self, name, id, sem):
        super().__init__(sem)
        self.name = name
        self.id = id

    def courses(self, *args):
        print(f"Courses Approved to this EEE student in {self.semester} semester :")
        for i in args:
            print(i)


s1 = CSE("Rahim", 16101328, "Spring2016")
s1.student_info()
s1.courses("CSE110", "MAT110", "ENG101")
print("==================")
s2 = EEE("Tanzim", 18101326, "Spring2018")
s2.student_info()
s2.courses("Mat110", "PHY111", "ENG101")
print("==================")
s3 = CSE("Rudana", 18101326, "Fall2017")
s3.student_info()
s3.courses("CSE111", "PHY101", "MAT120")
print("==================")
s4 = EEE("Zainab", 19201623, "Summer2019")
s4.student_info()
s4.courses("EEE201", "PHY112", "MAT120")
```
