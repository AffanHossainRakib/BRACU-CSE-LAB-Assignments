```python
#@title Task 2
class Assassin:
    n = 0

    def __init__(self, name, r=100):
        Assassin.n += 1
        self.name = name
        self.r = r

    def printDetails(self):
        print(f"Name: {self.name}")
        print(f"Success rate: {self.r}% ")
        print(f"Total number of Assassin: {Assassin.n}")

    @classmethod
    def failureRate(cls, name, r):
        return cls(name, 100 - r)

    @classmethod
    def failurePercentage(cls, name, r):
        return cls(name, 100 - int(r[0:2:1]))


# Write your code here
john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage("Akabane", "10%")
akabane.printDetails()
```

```python
#@title Task 5
class Employee:

    def __init__(self,name, wp):
        self.name = name
        self.workingPeriod = wp

    @staticmethod
    def experienceCheck(wp, gender):

        if gender == "male":
            pronoun = "He"
        else:
            pronoun = "She"

        if wp < 3:
            return f"{pronoun} is not experienced"
        else:
            return f"{pronoun} is experienced"

    @classmethod
    def employeeByJoiningYear(cls, name, workingPeriod):
        wp = 2023 - workingPeriod
        return cls(name, wp)


# Write your code here
employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
#@title Task 6
class Laptop:
    laptopCount = 0
    def __init__(self, name, count):
        self.name = name
        self.count = count
        Laptop.laptopCount += count

    @staticmethod
    def advantage():
        print("Laptops are portable")

    @classmethod
    def resetCount(cls):
        Laptop.laptopCount = 0

# Write your code here
lenovo = Laptop("Lenovo", 5);
dell = Laptop("Dell", 7);
print(lenovo.name, lenovo.count)
print(dell.name, dell.count)
print("Total number of Laptops", Laptop.laptopCount)
Laptop.advantage()
Laptop.resetCount()
print("Total number of Laptops", Laptop.laptopCount)
```

```python
#@title Task 7
class Cat:
    Number_of_cats = 0
    def __init__(self, c, p):
        Cat.Number_of_cats += 1
        self.c = c
        self.p = p

    def changeColor(self,c):
        self.c = c

    def printCat(self):
        print(f"{self.c} cat is {self.p}")
        
    @classmethod
    def no_parameter(cls):
        return cls("White","sitting")

    @classmethod
    def first_parameter(cls,c):
        return cls(c,"Sitting")

    @classmethod
    def second_parameter(cls, p):
        return cls("White", p)


print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#@title Task 8
import math


class Cylinder:
    radius = 5
    height = 18

    def __init__(self, r, h):
        print(f"Default radius={Cylinder.radius} and height={Cylinder.height}.")
        Cylinder.radius = r
        Cylinder.height = h
        print(f"Updated: radius={Cylinder.radius} and height={Cylinder.height}.")

        self.radius = Cylinder.radius
        self.height = Cylinder.height

    @staticmethod
    def area(r, h):
        res = 2 * math.pi * r * (r + h)
        print(f"Area: {res}")

    @staticmethod
    def volume(r, h):
        res = math.pi * math.pow(r, 2) * h
        print(f"Volume: {res}")

    @classmethod    
    def swap(cls, r, h):
        return cls(h, r)

    @classmethod
    def changeFormat(cls, x):
        r, h = x.split("-")
        r = float(r)
        h = float(h)
        return cls(r, h)


# Write your code here
c1 = Cylinder(0, 0)
Cylinder.area(c1.radius, c1.height)
Cylinder.volume(c1.radius, c1.height)
print("===============================")
c2 = Cylinder.swap(8, 3)
c2.area(c2.radius, c2.height)
c2.volume(c2.radius, c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius, c3.height)
c3.volume(c3.radius, c3.height)
print("===============================")
Cylinder(0.3, 5.56).area(Cylinder.radius, Cylinder.height)
print("===============================")
Cylinder(3, 5).volume(Cylinder.radius, Cylinder.height)
```

```python
#@title Task 10
class SultansDine:
    t_n_b = 0
    t_s = 0
    branch_list = []

    def __init__(self, name):
        self.name = name
        self.b_s = 0
        SultansDine.t_n_b += 1

    def branchInformation(self):
        print(f"Branch Name: {self.name}\nBranch Sell: {self.b_s} Taka")

    def sellQuantity(self, q):
        if q < 10:
            self.b_s = q * 300
        elif q < 20:
            self.b_s = q * 350
        else:
            self.b_s = q * 400

        SultansDine.t_s += self.b_s
        SultansDine.branch_list.append([self.name, self.b_s])

    @classmethod
    def details(cls):
        print(f"Total Number of branch(s): {SultansDine.t_n_b}\nTotal Sell: {SultansDine.t_s} Taka")
        for i in SultansDine.branch_list:
            print(f"Branch Name: {i[0]}, Branch Sell: {i[1]} Taka \nBranch consists of total sell's: {round(i[1] / SultansDine.t_s * 100, 2)}%")


SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
