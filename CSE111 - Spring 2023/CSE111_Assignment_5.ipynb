```python
#@title Task 2
class Triangle:
    def __init__(self, b, h):
        self.__b = b
        self.__h = h

    def getBase(self):
        return self.__b

    def getHeight(self):
        return self.__h

    def setBase(self, b):
        self.__b = b

    def setHeight(self, h):
        self.__h = h

    def area(self):
        return .5*self.__b*self.__h

t1 = Triangle(10, 5)
print("First Triangle Base:" , t1.getBase())
print("First Triangle Height:" , t1.getHeight())
print("First Triangle area:" ,t1.area())
t2 = Triangle(5, 3)
print("Second Triangle Base:" , t2.getBase())
print("Second Triangle Height:" , t2.getHeight())
print("Second Triangle area:" ,t2.area())

#Write this below part in the assignemnt
t3 = Triangle(5, 2)
print("Third Triangle Base:" , t3.getBase())
print("Third Triangle Height:" , t3.getHeight())
print("Third Triangle area:" ,t3.area())
```

```python
#@title Task 5
class Student:
    def __init__(self, name, id, dep):
        self.name = name
        self.id = id
        self.dep = dep
        self.email = None
        self.password = None
        self.login = None
        self.course = []
        print("Student object is created!")


class Usis:
    def __init__(self):
        print("USIS is ready to use!")

    def login(self, obj):
        if obj.email is not None and obj.password is not None:
            obj.login = True
            print("Login successful!")
        else:
            print("Email and password need to be set.")

    def advising(self, obj, *course):
        if obj.login is None:
            print("Please login to advise courses!")
        else:
            n = len(course)
            if n == 0:
                print("You haven't selected  any courses.")
            elif n > 3:
                print("You need special approval to take more than 3 courses.")
            else:
                print("Advising successful!")
                for i in course:
                    obj.course.append(i)

    def individualDetails(self, obj):
        res = ""
        for i in obj.course:
            res += i + ","

        return f"Name: {obj.name}\nID: {obj.id}\nDepartment: {obj.dep}\nAdvised courses: {res[0:-1:1]}"

rakib = Student("Rakib", 12301455, "CSE")
print("1***********************")
usis_obj = Usis()
print("2***********************")
usis_obj.login(rakib)
print("3***********************")
usis_obj.advising(rakib)
print("4***********************")
rakib.email = "rakib@hotmail.com"
rakib.password = "1234"
print("5***********************")
usis_obj.login(rakib)
print("6***********************")
usis_obj.advising(rakib)
print("7***********************")
usis_obj.advising(rakib, "CSE110", "PHY111", "MAT110","CSE260")
print("8***********************")
usis_obj.advising(rakib, "CSE110", "PHY111", "MAT110")
print("9***********************")
print(usis_obj.individualDetails(rakib))
```

```python
# @title Task 6
class Train:
    def __init__(self, name, *args):
        self.name = name
        self.journey = args
        self.fare = {}
        # self.place = args

        temp = 100
        for i in args:
            self.fare[i] = temp
            temp += 100

        self.p_data = []
        print(f"Welcome aboard on {self.name}")
        print(f"Start: {self.journey[0]}")
        print(f"Destination: {self.journey[-1]}")


    def addPassenger(self, *p):
        for i in p:
            print(f"{i.name} welcome aboard")

            if i.start is None:
                i.start = self.journey[0]
            if i.end is None:
                i.end = self.journey[-1]

            self.t = self.fare[i.end] - self.fare[i.start]
            self.p_data.append([i.name, i.start, i.end, self.t])

    def allPassengerDetails(self):
        for i in self.p_data:
            print(f"Name: {i[0]},Start: {i[1]},Destination:{i[2]},Fair: ${i[3]}")


class Passenger:
    def __init__(self, name, start=None, end=None):
        self.name = name
        self.start = start
        self.end = end


t1 = Train('T1-Express', 'New York', 'Manhattan', 'Brooklyn', 'Boston')
print("1========================")
p1 = Passenger("Naruto")
t1.addPassenger(p1)
p2 = Passenger("Sasuke", "Manhattan")
p3 = Passenger("Hinata", "Manhattan", "Brooklyn")
print("2========================")
t1.addPassenger(p2, p3)
print("3========================")
t1.allPassengerDetails()
print("4========================")
t2 = Train('Europe-Express', 'London', 'Paris', 'Brussels', 'Turkey')
print("5========================")
p4 = Passenger("Max", "London", "Brussels")
p5 = Passenger("Eleven", "Paris")
p6 = Passenger("Mike", "Brussels")
t2.addPassenger(p4, p5, p6)
print("6========================")
t2.allPassengerDetails()
```

```python
#@title Task 7
class BracuStudent:
    def __init__(self, name, home):
        self.name = name
        self.home = home
        self.b_pass = False

    def show_details(self):
        print(f"Studnet Name: {self.name}")
        print(f"Lives in {self.home}")
        print(f"Have Bus Pass? {self.b_pass}")

    def get_pass(self):
        self.b_pass = True


class BracuBus:
    def __init__(self, route, m=2):
        self.route = route
        self.p = 0
        self.m = m
        self.p_on_b = []

    def show_details(self):
        print(f"Bus Route: {self.route}")
        print(f"Passengers Count: {self.p} (Max: {self.m})")
        print(f"Passengers On Board: {self.p_on_b}")

    def board(self, *args):
        n = len(args)
        if n == 0:
            print("No passenger!")
        else:
            for obj in args:
                if obj.b_pass is True:
                    if obj.home is self.route:
                        if self.p < self.m:
                            print(f"{obj.name} boarded the bus")
                            self.p += 1
                            self.p_on_b.append(obj.name)
                        else:
                            print("Bus is full!")
                    else:
                        print("You got on wrong bus!")
                else:
                    print("You don't have bus pass!")


st1 = BracuStudent("Afif", "Mirpur")
print("1===========================")
st2 = BracuStudent("Shanto", "Motijheel")
st3 = BracuStudent("Taskin", "Mirpur")
st1.show_details()
st2.show_details()
print("2===========================")
st3.show_details()
print("3===========================")
bus1 = BracuBus("Mirpur")
bus2 = BracuBus("Azimpur", 5)
bus1.show_details()
bus2.show_details()
print("4===========================")
st2.get_pass()
st3.get_pass()
print("5===========================")
st2.show_details()
st3.show_details()
print("6===========================")
bus1.board()
print("7===========================")
bus1.board(st1, st2)
print("8===========================")
st1.get_pass()
st2.home = "Mirpur"
st1.show_details()
st2.show_details()
print("9===========================")
bus1.board(st1, st2, st3)
print("10===========================")
bus1.show_details()
```

```python
#@title Task 8
class Library:
    def __init__(self, name, books):
        self.name = name
        self.borrower = {}
        self.books = books

    def details(self):
        print(f"{self.name} Library details")
        print(f"Borrower details:\n{self.borrower}")
        print(f"Books availability:\n{self.books}")


class Reader:
    def __init__(self, name):
        self.name = name
        self.counter = 0
        self.borrowed = {}

    def borrow(self, obj, *args):
        if self.name not in obj.borrower:
            obj.borrower[self.name] = 0

        for i in args:
            if self.counter < 5:
                if obj.books[i] > 0:
                    print(f"{i} book is borrowed successfully.")
                    self.counter += 1
                    obj.borrower[self.name] += 1
                    obj.books[i] -= 1
                    if i not in self.borrowed:
                        self.borrowed[i] = 1
                    else:
                        self.borrowed[i] += 1
                else:
                    print(f"{i} books are not available at the moment.")
            else:
                print("You cannot borrow more than 5 books.")

    def readerInfo(self, genre=None):
        if genre is None:
            print(f"{self.name}, you have {self.counter} book(s) with you.")
            for i in self.borrowed:
                print(f"Books on {i}: {self.borrowed[i]}")
        else:
            print(f"{self.name}, you have {self.borrowed[genre]} {genre} book(s) with you.")


L1 = Library('Dhaka', {'Arts': 15, 'Fiction': 135, 'Politics': 2, 'Science': 11, 'Poetry': 15})
L1.details()
print("1----------------------")
r1 = Reader('Aladdin')
r1.borrow(L1, 'Arts', 'Fiction', 'Fiction', 'Politics')
print("2----------------------")
r1.borrow(L1, 'Politics', 'Fiction')
print("3----------------------")
r1.readerInfo()
print("4----------------------")
r1.readerInfo('Fiction')
print("5----------------------")
L1.details()
print("6----------------------")
r2 = Reader('Jasmine')
r2.borrow(L1, 'Politics', 'Poetry')
print("7----------------------")
r2.readerInfo()
print("8----------------------")
L1.details()
```
