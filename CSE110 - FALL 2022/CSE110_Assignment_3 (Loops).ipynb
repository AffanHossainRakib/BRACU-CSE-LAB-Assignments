```python
#@title Task 14
a = int(input())
sum = 0
for i in range(1,a+1,1):
    if a%i == 0 and i<a:
        sum = sum + i
if sum == a:
    print(a,'is a perfect number')
else:
    print(a,'is not a perfect number')
```

```python
#@title Task 16
quantity = int(input())
sum = 0
for i in range(quantity):
  usr=int(input())
  sum = sum + usr
  if i == 0:
    max = usr
    min = usr
  else:
    if usr > max:
      max = usr
    elif usr < min:
      min = usr
print('Maxium is', max)
print('Minimum is', min)
print('Average is', sum/quantity)
```
