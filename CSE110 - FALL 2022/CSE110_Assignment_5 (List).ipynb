```python
#@title Task 6
userInput = input()+','
myList = []
newstr = ''
max = 0
for i in userInput:
    if i != ',':
        newstr += i
    else:
        myList += [int(newstr)]
        newstr = ''
print('My list:',myList)
for i in myList:
    if i > max:
        max = i
print('Largest number in the list is', max ,'which was found at index', str(myList.index(max))+'.')
```

```python
#@title Task 9
userInput = input()+' '
originalList = []
newstr = ''
modifiedList = []

for i in userInput:
    if i != ' ':
        newstr += i
    else: 
       originalList += [int(newstr)] 
       newstr = ''
print('Original list:',originalList)

for j in originalList:
    if j%2 != 0:
        modifiedList += [j]
print('Modified list:',modifiedList)
```

```python
#@title Task 10
userInput = input()+','
newstr = ''
inputList = []
modifiedList = []
for i in userInput:
    if i != ',':
        newstr += i
    else:
        inputList+= [int(newstr)]
        newstr = ''
print('Input list:',inputList)

for j in inputList:
    if j not in modifiedList:
        modifiedList += [j]
print('Modified list:',modifiedList)
```
