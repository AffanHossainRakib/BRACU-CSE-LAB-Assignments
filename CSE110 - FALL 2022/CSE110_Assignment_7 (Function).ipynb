```python
#Task 4
def upLowCounter(var):
    cap = 0
    low = 0
    for i in var:
        if 65 <= ord(i) <= 90:
            cap += 1
        elif 97 <= ord(i) <= 122:
            low += 1
    print(f"No. of Uppercase characters: {cap}")
    print(f"No. of Lowercase Characters: {low}")
upLowCounter('HaRRy PotteR')
```

```python
# Task 8
def show_palindrome(fx):
    mystr = ''
    for i in range(1,fx+1):
        mystr += str(i)
    newstr = mystr[-2::-1]
    return mystr+newstr

def show_palindromic_triangle(n):
    for i in range(1, n+1, 1):
        palindrome = show_palindrome(i)
        for j in range(0, n-i):
            print(" ",end=" ")
        for x in palindrome:
            print(x,end = " ")
        print()
show_palindromic_triangle(6)
```

```python
#Task 9
def area_circumference_generator(radius):
    import math
    area = math.pi*radius**2
    circumference = 2*math.pi*radius
    return area,circumference
rvar = area_circumference_generator(1)
area,circumference = rvar
print(rvar)
print(f"Area of the circle is {area} and circumference is {circumference}")
```

```python
# Task 11
def rem_duplicate(var):
    myList = []
    for i in var:
        if i not in myList:
            myList.append(i)
    return (tuple(myList))
x = rem_duplicate(("Hi", 1, 2, 3, 3, "Hi",'a', 'a', [1,2]))
print(x)
```

```python
# Task 12
def myRemover(var):
    myList = []
    counter = 0
    for i in var:
        if myList.count(i) == 0 or myList.count(i) == 1:
            myList.append(i)
        else:
            counter += 1
    print(f"Removed: {counter}")
    return myList
x = myRemover([1, 2, 3, 3, 3, 3, 4, 5, 8, 8])
print(x)
```
