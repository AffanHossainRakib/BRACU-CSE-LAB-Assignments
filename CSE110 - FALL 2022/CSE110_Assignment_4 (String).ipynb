```python
#@title Task 10
var = input()
var1 = ''
for i in var:
    if i == ',':
        break
    var1 = var1 + i
lv1 = len(var1)
var2 = var[lv1+2::]
var3 = ''
lv1 = len(var1)
lv2 = len(var2)
if lv1 >= lv2:
    for i in range(lv2):
        var3 = var3+var1[i]+var2[i]
elif lv1 < lv2:
    for i in range(lv1):
        var3 = var3+var1[i]+var2[i]
if lv1 > lv2:
    var3 = var3+var1[lv2:lv1:1]
elif lv1 < lv2:
    var3 = var3+var2[lv1:lv2:1]
print(var3)
```
