```python
# Test 01: Play Right
def playRight(sequence, beats):
    i = 0
    while i < len(beats):
        if beats[i] == 1:
            rightRotate(sequence)
        i += 1

    return sequence


def rightRotate(arr):
    temp = arr[len(arr)-1]
    i = len(arr) - 1
    while i > 0:
        arr[i] = arr[i-1]
        i -= 1
    arr[0] = temp


print("///  Test 01: Play Right  ///")
sequence = np.array([10, 20, 30, 40, 50, 60])
beats = np.array([1, 0, 0, 1, 0, 1])
returned_value = playRight(sequence, beats)
# This should print [40, 50, 60, 10, 20, 30]
print(f'Task 1: {returned_value}')
```

```python
# Test 02: Discard Cards
def discardCards(cards, number):
    i = 0
    while i < len(cards):
        if cards[i] == number:
            leftShift(cards, i)
            i -= 1
        i += 1

    return cards


def leftShift(arr, index):
    while index < len(arr) - 1:
        arr[index] = arr[index+1]
        index += 1
    arr[len(arr)-1] = 0


print("///  Test 02: Discard Cards  ///")
cards = np.array([1, 2, 3, 2, 8, 2, 2, 5, 7])
returned_value = discardCards(cards, 2)
# This should print [1, 3, 8, 5, 7, 0, 0, 0, 0]
print(f'Task 2: {returned_value}')
unittest.output_test(returned_value, np.array([1, 3, 8, 5, 7, 0, 0, 0, 0]))
```

```python
# Test 03: Merge Lineup
def mergeLineup(pokemon_1, pokemon_2):
    result = [None]*len(pokemon_1)  # this array stores the resulting array

    i = 0
    j = len(pokemon_1) - 1

    while i <= j:
        if pokemon_1[i] == None:
            pokemon_1[i] = 0
        if pokemon_2[j-i] == None:
            pokemon_2[j-i] = 0

        result[i] = pokemon_1[i] + pokemon_2[j-i]

        i += 1

    return np.array(result)


print("///  Test 03: Merge Lineup  ///")
pokemon_1 = np.array([12, 3, 25, 1, None])
pokemon_2 = np.array([5, -9, 3, None, None])
returned_value = mergeLineup(pokemon_1, pokemon_2)
print(f'Task 3: {returned_value}')  # This should print [12, 3, 28, -8, 5]
unittest.output_test(returned_value, np.array([12, 3, 28, -8, 5]))

pokemon_1 = np.array([4, 5, -1, None, None])
pokemon_2 = np.array([2, 27, 7, 12, None])
returned_value = mergeLineup(pokemon_1, pokemon_2)
print(f'Task 3: {returned_value}')  # This should print [4,17,6,27,2]
unittest.output_test(returned_value, np.array([4, 17, 6, 27, 2]))
```

```python
# Test 04: Balance your Salami
def balanceSalami(salami):
    total = 0
    i = 0
    while i < len(salami):
        total += salami[i]
        i += 1

    half = 0
    i = 0
    while i < len(salami):
        half += salami[i]
        if total - half == half:
            return True
        i += 1

    return False


print("///  Test 04: Balance Your Salami  ///")
salami = np.array([1, 1, 1, 2, 1])
returned_value = balanceSalami(salami)
print(f'Task 4: {returned_value}')  # This should print True
unittest.output_test(returned_value, True)

salami = [2, 1, 1, 2, 1]
returned_value = np.array(balanceSalami(salami))
print(f'Task 4: {returned_value}')  # This should print False
unittest.output_test(returned_value, False)

salami = [10, 3, 1, 2, 10]
returned_value = np.array(balanceSalami(salami))
print(f'Task 4: {returned_value}')  # This should print True
unittest.output_test(returned_value, True)
```

```python
# Test 05: Protecc Salami
def protectSalami(salami):
    frequncey = np.array([-1]*len(salami))
    i = 0
    while i < len(salami):
        if frequncey[i] == -1:
            frequncey[i] = 0
            j = i + 1
            while j < len(salami):
                if salami[j] == salami[i]:
                    frequncey[i] += 1
                    frequncey[j] = 0
                j += 1
        i += 1

    i = 0
    while i < len(frequncey):
        j = i + 1
        if frequncey[i] != 0:
            while j < len(frequncey):
                if frequncey[i] == frequncey[j]:
                    return True
                j += 1
        i += 1

    return False


print("///  Test 05: Protecc Salami  ///")
salami = np.array([4, 5, 6, 6, 4, 3, 6, 4])
returned_value = protectSalami(salami)
print(f'Task 5: {returned_value}')  # This should print True
unittest.output_test(returned_value, True)

salami = np.array([3, 4, 6, 3, 4, 7, 4, 6, 8, 6, 6])
returned_value = protectSalami(salami)
print(f'Task 5: {returned_value}')  # This should print False
unittest.output_test(returned_value, False)
```

```python
# Playing with Array and Probability Theory
# Lab1_Part2_Task

def mean_value(arr):
    n = len(arr)
    sumation = 0
    i = 0
    while i < n:
        sumation += arr[i]
        i += 1

    mean = sumation/n
    return mean


def std_deviation(arr):
    n = len(arr)
    mean = mean_value(arr)
    deviation = 0
    for i in range(n):
        deviation += (arr[i] - mean) ** 2

    deviation = (deviation / (n - 1)) ** .5

    return deviation


def std_dev_away(arr):
    n = len(arr)
    mean = mean_value(arr)
    dev = std_deviation(arr)

    counter = 0
    for i in range(n):
        if mean + dev*1.5 < arr[i]:
            counter += 1
        elif mean - dev*1.5 > arr[i]:
            counter += 1

    new_arry = np.array([None]*counter)

    counter = 0
    for i in range(n):
        if mean + dev*1.5 <= arr[i]:
            new_arry[counter] = arr[i]
            counter += 1
        elif mean - dev*1.5 >= arr[i]:
            new_arry[counter] = arr[i]
            counter += 1

    return new_arry


# Driver Code
arr = np.array([10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4])

returned_value = mean_value(arr)
print(f"The mean of the numbers is: {round(returned_value,12)}")

returned_value = std_deviation(arr)
print(f"The standard deviation is: {round(returned_value,2)}")

returned_value = std_dev_away(arr)
print(f"New array: {returned_value}")
```
