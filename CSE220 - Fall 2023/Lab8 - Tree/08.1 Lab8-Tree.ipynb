```python
class BTNode:
  def __init__(self, elem):
    self.elem = elem
    self.right = None
    self.left = None
```

```python
def inorder(root):
  if root == None:
    return

  inorder(root.left)
  print(root.elem, end = ' ')
  inorder(root.right)
```

```python
def tree_construction(arr, i = 1):
  if i>=len(arr) or arr[i] == None:
    return None
  p = BTNode(arr[i])
  p.left = tree_construction(arr, 2*i)
  p.right = tree_construction(arr, 2*i+1)
  return p


root2 = tree_construction([None, 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', None, None, None, 'I', 'J', None, 'k'])
inorder(root2)
```

```python
def convert_mirror(root):
    if not root:
        return
    
    new_root = BTNode(root.elem)

    new_root.left = convert_mirror(root.right)
    new_root.right = convert_mirror(root.left)

    return new_root

# DRIVER CODE
root = BTNode(10)
n1 = BTNode(20)
n2 = BTNode(30)
n3 = BTNode(40)
n4 = BTNode(60)

root.left = n1
root.right = n2

n1.left = n3
n1.right = n4

print("Given Tree Inorder Traversal: ", end=" ")
inorder(root)  # Given Tree Inorder Traversal:  40 20 60 10 30
print()

root2 = convert_mirror(root)
print("Mirrored Tree Inorder Traversal: ", end=" ")
inorder(root2)  # Mirrored Tree Inorder Traversal:  30 10 60 20 40
```

```python
def smallest_level(root, curr_lvl = 0, lvl_dir={}):
    if not root:
        return 
    
    if curr_lvl not in lvl_dir:
        lvl_dir[curr_lvl] = root.elem
    else:
        if lvl_dir[curr_lvl] > root.elem:
            lvl_dir[curr_lvl] = root.elem
    smallest_level(root.left, curr_lvl + 1, lvl_dir)
    smallest_level(root.right, curr_lvl + 1, lvl_dir)

    return lvl_dir

# DRIVER CODE
root = tree_construction([None, 4, 9, 2, 3, -5, None, 7])
print("Given Tree Inorder Traversal: ", end=" ")
inorder(root)  # Given Tree Inorder Traversal:  3 9 -5 4 2 7
print()
print("Level Wise Smallest Value: ", end=" ")
print(smallest_level(root))  # Level Wise Smallest Value:  {0: 4, 1: 2, 2: -5}
```

```python
def sub_tree_max(root):
    tail = root
    while not tail.right:
        tail = tail.right
    return tail
    
def inorder_predecessor(root, x, mother_root=None):
    if root == None:
        return
    
    if x.elem < root.elem:
        return inorder_predecessor(root.left, x, mother_root)
    elif x.elem > root.elem:
        return inorder_predecessor(root.right, x, root)
    elif x.elem == root.elem:
        if root.left != None:
            return sub_tree_max(root.left)
        else:
            return mother_root

# DRIVER CODE
root = BTNode(20)
n1 = BTNode(8)
n2 = BTNode(22)
n3 = BTNode(4)
n4 = BTNode(12)
n5 = BTNode(10)
n6 = BTNode(14)

root.left = n1
root.right = n2

n1.left = n3
n1.right = n4

n4.left = n5
n4.right = n6

print("Given Tree Inorder Traversal: ", end=" ")
inorder(root)  # Given Tree Inorder Traversal:  4 8 10 12 14 20 22
print()

x = root
print(f"Inorder predecessor of node {x.elem}: {inorder_predecessor(root, x).elem}")  # Inorder predecessor of node 20: 14
```

```python
def LCA(root, x, y):
    if root == None:
        return
    
    if x < root.elem and y < root.elem:
        return LCA(root.left, x, y)
    elif x > root.elem and y > root.elem:
        return LCA(root.right, x, y)
    return root.elem


# DRIVER CODE
root = BTNode(15)

root.left = BTNode(10)
root.right = BTNode(25)

root.left.left = BTNode(8)
root.left.right = BTNode(12)

root.left.left.left = BTNode(6)
root.left.left.right = BTNode(9)

root.right.left = BTNode(20)
root.right.right = BTNode(30)

root.right.left.left = BTNode(18)
root.right.left.right = BTNode(22)

print("Tree Inorder Traversal: ", end=" ")
inorder(root)  # Given Tree Inorder Traversal: 6 8 9 10 12 15 18 20 22 25 30
print()

x, y = 6, 10
print(f"Lowest Common Ancestor (LCA) of the pair {x,y} nodes in this BST is : {LCA(root,x,y)}")
x, y = 20, 6
print(f"Lowest Common Ancestor (LCA) of the pair {x,y} nodes in this BST is : {LCA(root,x,y)}")
x, y = 18, 22
print(f"Lowest Common Ancestor (LCA) of the pair {x,y} nodes in this BST is : {LCA(root,x,y)}")
x, y = 20, 25
print(f"Lowest Common Ancestor (LCA) of the pair {x,y} nodes in this BST is : {LCA(root,x,y)}")
x, y = 10, 12
print(f"Lowest Common Ancestor (LCA) of the pair {x,y} nodes in this BST is : {LCA(root,x,y)}")
```

```python
def SumTree(root):
    if not root:
        return  0
    
    left_sum = SumTree(root.left)
    right_sum = SumTree(root.right)

    return root.elem + left_sum + right_sum  

# DRIVER CODE
root = BTNode(26)

root.left = BTNode(10)
root.right = BTNode(3)

root.left.left = BTNode(4)
root.left.right = BTNode(6)

root.right.right = BTNode(3)

print("Tree Inorder Traversal: ", end=" ")
inorder(root)  # Given Tree Inorder Traversal: 4 10 6 26 3 3
print()

print(root.elem == SumTree(root)//2)
```
