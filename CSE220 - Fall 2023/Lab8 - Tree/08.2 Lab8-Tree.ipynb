```python
class BTNode:
    def __init__(self, elem):
        self.elem = elem
        self.right = None
        self.left = None
```

```python
def inorder(root):
    if root == None:
        return

    inorder(root.left)
    print(root.elem, end=" ")
    inorder(root.right)
```

```python
def tree_construction(arr, i=1):
    if i >= len(arr) or arr[i] == None:
        return None
    p = BTNode(arr[i])
    p.left = tree_construction(arr, 2 * i)
    p.right = tree_construction(arr, 2 * i + 1)
    return p


root2 = tree_construction([None, 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', None, None, None, 'I', 'J', None, 'k'])
inorder(root2)
```

```python
def convert_mirror(root):
    if root == None:
        return None
    node = BTNode(root.elem)
    node.left = convert_mirror(root.right)
    node.right = convert_mirror(root.left)

    return node


#DRIVER CODE
root = BTNode(10)
n1 = BTNode(20)
n2 = BTNode(30)
n3 = BTNode(40)
n4 = BTNode(60)

root.left = n1
root.right = n2

n1.left = n3
n1.right = n4

print('Given Tree Inorder Traversal: ', end = ' ')
inorder(root) #Given Tree Inorder Traversal:  40 20 60 10 30
print()

root2 = convert_mirror(root)
print('Mirrored Tree Inorder Traversal: ', end = ' ')
inorder(root2) #Mirrored Tree Inorder Traversal:  30 10 60 20 40
```

```python
def smallest_level(root):
    return helper_1(root, {}, 0)

def helper_1(root, dict, lvl):
    if root == None:
        return

    if lvl not in dict:
        dict[lvl] = root.elem
    else:
        if dict[lvl] > root.elem:
            dict[lvl] = root.elem

    helper_1(root.left, dict, lvl + 1)
    helper_1(root.right, dict, lvl + 1)
    return dict

#DRIVER CODE
root = tree_construction([None, 4,9,2,3,-5,None,7])
print('Given Tree Inorder Traversal: ', end = ' ')
inorder(root) #Given Tree Inorder Traversal:  3 9 5 4 2 7
print()
print('Level Wise Smallest Value: ', end = ' ')
print(smallest_level(root)) #Level Wise Smallest Value:  {0: 4, 1: 2, 2: -5}
```

```python
def inorder_predecessor(root, x):
    return helper_2(root, x)


def maximum(root):
    if root.right == None:
        return root
    return maximum(root.right)


def helper_2(root, x, m_root=None):
    if root == None:
        return

    if x.elem > root.elem:
        return helper_2(root.right, x, root)
    elif x.elem < root.elem:
        return helper_2(root.left, x, m_root)
    else:
        if root.left != None:
            return maximum(root.left)
        else:
            return m_root



#DRIVER CODE
root = BTNode(20)
n1 = BTNode(8)
n2 = BTNode(22)
n3 = BTNode(4)
n4 = BTNode(12)
n5 = BTNode(10)
n6 = BTNode(14)

root.left = n1
root.right = n2

n1.left = n3
n1.right = n4

n4.left = n5
n4.right = n6

print('Given Tree Inorder Traversal: ', end = ' ')
inorder(root) #Given Tree Inorder Traversal:  4 8 10 12 14 20 22
print()

x = root
print(f'Inorder predecessor of node {x.elem}: {inorder_predecessor(root, x).elem}') #Inorder predecessor of node 20: 14
```

```python
def LCA(root, x, y):
    if root == None:
        return

    if x < root.elem and y < root.elem:
        return LCA(root.left, x, y)
    elif x > root.elem and y > root.elem:
        return LCA(root.right, x, y)
    else:
        return root.elem

#DRIVER CODE
#Write by yourself from the given tree
#check all the sample inputs given
root = BTNode(15)

root.left = BTNode(10)
root.right = BTNode(25)

root.left.left = BTNode(8)
root.left.right = BTNode(12)

root.left.left.left = BTNode(6)
root.left.left.right = BTNode(9)

root.right.left = BTNode(20)
root.right.right = BTNode(30)

root.right.left.left = BTNode(18)
root.right.left.right = BTNode(22)

print("Tree Inorder Traversal: ", end=" ")
inorder(root)  # Given Tree Inorder Traversal: 6 8 9 10 12 15 18 20 22 25 30
print()

x, y = 6, 10
print(f"Lowest Common Ancestor (LCA) of the pair {x,y} nodes in this BST is : {LCA(root,x,y)}")
x, y = 20, 6
print(f"Lowest Common Ancestor (LCA) of the pair {x,y} nodes in this BST is : {LCA(root,x,y)}")
x, y = 18, 22
print(f"Lowest Common Ancestor (LCA) of the pair {x,y} nodes in this BST is : {LCA(root,x,y)}")
x, y = 20, 25
print(f"Lowest Common Ancestor (LCA) of the pair {x,y} nodes in this BST is : {LCA(root,x,y)}")
x, y = 10, 12
print(f"Lowest Common Ancestor (LCA) of the pair {x,y} nodes in this BST is : {LCA(root,x,y)}")
```

```python
#WRITE THE FUNCTION and DRIVER CODE BY YOURSELF
def SumTree(root):
    if root.elem == helper_3(root.left) + helper_3(root.right):
        return True
    return False

def helper_3(node):
    if node == None:
        return 0
    left_sum = helper_3(node.left)
    right_sum = helper_3(node.right)
    return node.elem + left_sum + right_sum

# DRIVER CODE
root = BTNode(26)

root.left = BTNode(10)
root.right = BTNode(3)

root.left.left = BTNode(4)
root.left.right = BTNode(6)

root.right.right = BTNode(3)

print("Tree Inorder Traversal: ", end=" ")
inorder(root)  # Given Tree Inorder Traversal: 4 10 6 26 3 3
print()

print(SumTree(root))
```
