```python
class Patient:
  def __init__(self, id, name, age, bloodgroup):
    self.id = id
    self.name = name
    self.age = age
    self.bloodgroup = bloodgroup
    self.next = None
    self.prev = None

class WRM:
  def __init__(self):
    temp = Patient(None, None, None, None)
    temp.next = temp
    temp.prev = temp

    self.head = temp
    self.curr = self.head

  def RegisterPatient(self, id, name, age, bloodgroup):
    self.curr.next = Patient(id, name, age, bloodgroup)
    self.curr.next.prev = self.curr
    self.curr = self.curr.next
    self.curr.next = self.head
    self.head.prev = self.curr
    print()
    print("Patient added successfully.")
    print()

  def ServePatient(self):
    temp = self.head.next
    if temp == self.head:
      print()
      print("No Patients Left To Serve.")
      print()
    else:
      print(f"{temp.name} served.")
      self.head.next = temp.next
      temp.next.prev = self.head


  def CancelAll(self):
    temp = self.head.next
    if temp == self.head:
      print()
      print("No appointments left.")
      print()
    else:
      temp = Patient(None, None, None, None)
      temp.next = temp
      temp.prev = temp
      self.head = temp
      self.curr = self.head
      print()
      print("All appointments cancelled successfully.")
      print()


  def CanDoctorGoHome(self):
    return self.head.next == self.head


  def ShowAllPatient(self):
    temp = self.head.next
    if temp != self.head:
      print()
      print('All patients:', end = " ")
      while temp != self.head:
        print(temp.name, end = " ")
        temp = temp.next
    else:
      print(f"No patients available.")

    print()


  def ReverseTheLine(self):
    temp = self.head.prev
    curr = temp.next

    if temp == self.head:
      print()
      print("No patients left to reverse.")
      print()
    else:
      while temp != self.head:
        temp.next = temp.prev
        temp.prev = curr
        temp = temp.next
        curr = temp.next
        if temp == self.head:
          temp.next = temp.prev
          temp.prev = curr
          break
      self.head = temp
      print()
      print("Line reversed successfully.")
      print()



print("<---- Welcome to Waiting Room Management (WRM) System --->")
print()
waiting_room = WRM()

while 1:
    print("==============================")
    print("1. Register Patient")
    print("2. Serve Patient")
    print("3. Show All Patient")
    print("4. Can Doctor Go Home?")
    print("5. Cancel All Appointments")
    print("6. Reverse the line")
    print("0. Exit")
    print("==============================")

    option = int(input("Choose an option to proceed: "))

    if option == 1:
        id = int(input("Enter ID: "))
        name = input("Enter Name: ")
        age = int(input("Enter Age: "))
        bloodgroup = input("Enter Blood Group: ")
        waiting_room.RegisterPatient(id, name, age, bloodgroup)

    elif option == 2:
        waiting_room.ServePatient()

    elif option == 3:
        waiting_room.ShowAllPatient()

    elif option == 4:
        flag = waiting_room.CanDoctorGoHome()

        if flag:
            print("Yes, doctor can go home")
        else:
            print("No, doctor cannot go home")

    elif option == 5:
        waiting_room.CancelAll()

    elif option == 6:
        waiting_room.ReverseTheLine()

    elif option == 0:
        break

    else:
        print('Enter valid number: ')
```
